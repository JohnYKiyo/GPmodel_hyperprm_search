```python
from GPRegression_boston import test, train, load_data
```

```python
X_train, X_test, Y_train, Y_test = load_data()
gpr = train(0.1, 0.1, X_train, Y_train)
RMSE = test(gpr, X_test, Y_test)
print(RMSE)
```

```python
import numpy as np
import itertools
```

```python
scale = np.logspace(-3, 3, 100)
alpha = np.logspace(-3, 3, 100)
```

```python
loss = []
sc_axis = []
al_axis = []
for sc,al in itertools.product(scale,alpha):
    gpr = train(sc, al, X_train, Y_train)
    RMSE = test(gpr, X_test, Y_test)
    loss.append(RMSE)
    sc_axis.append(sc)
    al_axis.append(al)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.contourf(np.array(sc_axis).reshape(100,100), 
             np.array(al_axis).reshape(100,100), 
             np.array(loss).reshape(100,100),
             levels=10)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('scale')
plt.ylabel('alpha')
plt.show()
```
